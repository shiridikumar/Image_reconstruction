```python
supports=[1000,10000,20000,50000,100000]
numpatterns=[]
runtimes=[]
memory=[]
for sup in supports:
    miner=alg.PositionMining(sup,"D3.csv")
    miner.startMine()
    allpatterns=miner.getPatterns()
    numpatterns.append(len(allpatterns))
    runtimes.append(miner.getRuntime())
    memory.append(miner.getMemoryRSS())
    print("Support ", sup,"finsished")
# print(miner.getPatterns())
```

```python
import matplotlib.pyplot as plt
plt.title("Runtime Vs Support count")
plt.xlabel("Min Support count")
plt.ylabel("Runtime")
plt.plot(supports,runtimes)
plt.show()
```

```python
plt.title("Number of frequent patterns Vs Support count")
plt.xlabel("Min Support count")
plt.ylabel("Number of Frequent Patterns")
plt.plot(supports,numpatterns)
```

```python
plt.title("Memory Usage Vs Support counts")
plt.plot(supports,memory)
```
